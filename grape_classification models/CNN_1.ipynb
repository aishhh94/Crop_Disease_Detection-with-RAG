```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
X_train = np.load("data/X_train.npy")
y_train = np.load("data/y_train.npy")
X_val = np.load("data/X_val.npy")
y_val = np.load("data/y_val.npy")
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Input(shape=(224,224,3)))
model.add(tf.keras.layers.Conv2D(32,(3,3),activation="relu"))
model.add(tf.keras.layers.MaxPooling2D((2,2)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128,activation="relu"))
model.add(tf.keras.layers.Dense(4,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor="val_loss",patience=3,restore_best_weights=True)
```

```python
model.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=20,batch_size=32,callbacks=[early_stop])
```

```python
from sklearn.metrics import classification_report

pred = model.predict(X_val)
pred_classes = pred.argmax(axis=1)

print(classification_report(y_val, pred_classes, target_names=["Grape___Black_rot","Grape___Esca_(Black_Measles)","Grape___healthy","Grape___Leaf_blight_(Isariopsis_Leaf_Spot)"]))
```
